```python
# Load MVP winners data
mvp_df = pd.read_csv("data/raw/mvp_winners.csv")

# Show the first few rows
mvp_df.head()
```

```python
# Convert player names to uppercase and remove extra spaces
mvp_df["Player"] = mvp_df["Player"].str.upper().str.strip()
df["PLAYER"] = df["PLAYER"].str.upper().str.strip()
```

```python
# Merge player stats with MVP data based on player name
merged_df = df.merge(mvp_df, left_on="PLAYER", right_on="Player", how="left")

# Create MVP Label: If player is in MVP list, set MVP = 1, else 0
merged_df["MVP"] = merged_df["Year"].notna().astype(int)

# Drop unnecessary columns
merged_df.drop(columns=["Year", "Player", "Team"], inplace=True)

# Show the first few rows
merged_df.head()
```

```python
merged_df.to_csv("data/processed/cleaned_data.csv", index=False)

print("✅ Processed data saved to data/processed/cleaned_data.csv")
```
